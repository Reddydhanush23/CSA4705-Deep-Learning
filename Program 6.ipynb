```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_path = "/content/deeplearning 6th.jpg"  # your local image path
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
if img is None:
    raise ValueError(" Image not found! Check the file path or filename.")
blur = cv2.GaussianBlur(img, (5, 5), 0)
ret1, th1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret2, th2 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(f"Manual threshold value used: 120")
print(f"Otsu's optimal threshold value: {ret2:.2f}")
titles = ['Original Image', 'Manual Thresholding', "Otsuâ€™s Thresholding"]
images = [img, th1, th2]
plt.figure(figsize=(12, 6))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
plt.figure(figsize=(6,4))
plt.hist(img.ravel(), 256, [0, 256])
plt.axvline(x=ret2, color='red', linestyle='--', label=f"Otsu Threshold = {ret2:.2f}")
plt.title("Image Histogram with Otsu's Threshold")
plt.legend()
plt.show()
```
