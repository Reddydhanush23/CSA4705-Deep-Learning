```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2

def df(x):
    return 2*x
x = 8
alpha = 0.1
iterations = 15
x_vals = [x]
y_vals = [f(x)]
for i in range(iterations):
    x = x - alpha * df(x)
    x_vals.append(x)
    y_vals.append(f(x))
x_range = np.linspace(-10, 10, 200)
plt.plot(x_range, f(x_range), label="f(x) = xÂ²")
plt.scatter(x_vals, y_vals, color='red')
plt.plot(x_vals, y_vals, 'ro--', label="Descent Path")
plt.title("Gradient Descent Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
